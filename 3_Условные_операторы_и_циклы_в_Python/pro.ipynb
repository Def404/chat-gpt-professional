# Задание 1

Напишите программу по оценке экзаменационных результатов пользователей на основе введенных баллов (максимум 100 баллов).

1. Если пользователь набрал 90 баллов и выше - получает оценку "Отлично".
2. Оценка "Хорошо" начинается от 70 баллов (включительно).
3. Оценка "Удовлетворительно" от 50 баллов (включительно).
4. Если оценка ниже - то экзамен считается проваленным, предложите пользователю прийти на пересдачу.

Необходимо реализовать проверку введенных пользователем данных, что это числа, а не буквы, и что числа находятся в диапазоне от 0 до 100.

```python
score = input("Введите количество баллов: ")

if score.isdigit():
    score = int(score)
    if score >= 90 and score <= 100:
        print("Оценка: Отлично")
    elif score >= 70 and score < 90:
        print("Оценка: Хорошо")
    elif score >= 50 and score < 70:
        print("Оценка: Удовлетворительно")
    elif score >= 0 and score < 50:
        print("Экзамен провален. Рекомендуем прийти на пересдачу.")
    else:
      print("Ошибка: Введите число в диапазоне от 0 до 100.")
else:
   print("Ошибка: Введите числовое значение.")
```

# Задание 2

Выполните задание 1, но с применение pattern matching

```python
score = input("Введите количество баллов: ")

if score.isdigit():
    score = int(score)
    match score:
        case _ if 90 <= score <= 100:
            print("Оценка: Отлично")
        case _ if 70 <= score < 90:
            print("Оценка: Хорошо")
        case _ if 50 <= score < 70:
            print("Оценка: Удовлетворительно")
        case _ if 0 <= score < 50:
            print("Экзамен провален. Рекомендуем прийти на пересдачу.")
        case _:
            print("Ошибка: Введите число в диапазоне от 0 до 100.")
else:
    print("Ошибка: Введите числовое значение.")
```

# Задание 3

Машины по очереди заезжают на парковку, пока в ней есть свободные места. Вместимость парковки 8 авто.

Используя цикл `while`, каждый заезд очередной машины сопроводите сообщением `"Машина заехала"`. Выведите на экран количество свободных мест на парковке после заезда очередной машины.

Когда парковка окажется полной - сообщите остальным автовладельцам, что места в ней больше нет

```python
max_car_in_park = 8
count_car_in_park = 0

while True:
  if count_car_in_park < max_car_in_park:
    print('Машина заехала')
    count_car_in_park += 1
  else:
    print('Места на парковке больше нет')
    break



```
