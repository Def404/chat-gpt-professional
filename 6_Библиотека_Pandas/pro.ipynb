```python
data = {
    'Наименование товара': ['Телевизор', 'Телефон', 'Кофемолка', 'Наушники', 'Пылесос'],
    'Цена': [22022.43, 62021.34, 42020.66, 12022.24, 20241.12],
    'Остаток': [1, 3, 7, 2, 4]
}

df3 = pd.DataFrame(data)

display(df3)
```
